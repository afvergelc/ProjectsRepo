# **3. ANÁLISIS EXPLORATORIO DE DATOS**

## **3.1 Descripcion de las columnas o variables**

El dataset proviene de un hogar individual ubicado en Sceaux, Francia, y contiene mediciones del consumo eléctrico registradas entre diciembre de 2006 y noviembre de 2010. Los datos fueron recolectados a intervalos de un minuto, lo que proporciona una visión detallada del uso energético a lo largo del tiempo.

**Características del Dataset**
El dataset tiene un total de 2,075,259 registros y 9 columnas. Las cuales contienen:

**1. Date:** Fecha de la medición en formato dd/mmmm/aaaa.
Indica el día y mes de la medición.

**2. Time:** Hora de la medición en formato hh:mm:ss.
Proporciona el momento exacto en el que se realizó la medición.

**3. Global_active_power:** Potencia activa global promedio por minuto (en kilovatios). Representa el consumo energético principal del hogar y no considera los electrodomésticos que se encuentran en las siguientes columnas. Se encuentra asociado a la potencia que se puede emplear.

**4. Global_reactive_power:** Potencia reactiva global promedio por minuto (en kilovatios). Es el consumo energético que puede cargar un sistema y se encuentra almacenado al interior del mismo.
La potencia reactiva es crucial para el funcionamiento de dispositivos que requieren campos electromagnéticos, como motores o transformadores. Aunque no realiza trabajo útil directamente, es necesaria para mantener la estabilidad del sistema eléctrico.

**5. Voltage:** Voltaje promedio por minuto (en voltios).
El voltaje indica la presión eléctrica en el sistema, y su estabilidad es esencial para el correcto funcionamiento de los electrodomésticos. Fluctuaciones en el voltaje pueden afectar la eficiencia del consumo energético y el rendimiento de los dispositivos, lo que hace importante su monitoreo para detectar posibles problemas en la red eléctrica o en la infraestructura interna.

**6. Global_intensity:** Intensidad global promedio por minuto (en amperios).
La intensidad refleja la cantidad de corriente eléctrica utilizada por los dispositivos conectados en el hogar. Es directamente proporcional a la potencia activa y reactiva, permitiendo entender el comportamiento del consumo en términos de corriente eléctrica y, por ende, la carga total en el sistema.

**7. Sub_metering_1:** Submedición de energía activa en la cocina (en vatios-hora).
Esta submedición indica el consumo energético específico de los electrodomésticos utilizados en la cocina, como la nevera, estufas o microondas.

**8. Sub_metering_2:** Submedición de energía activa en la lavandería (en vatios-hora).
Refleja el consumo energético de los electrodomésticos en la lavandería, como lavadoras y secadoras.

**9. Sub_metering_3:** Submedición de energía activa asociada al calentador de agua eléctrico y aire acondicionado (en vatios-hora).
Mide el consumo energético de dispositivos de alto consumo como el calentador de agua eléctrico y el aire acondicionado.

Se desean validar patrones o tendencias.

```python
df = df.set_index('DateTime')

# Promedio mensual de Global Active Power (en kW)
df_monthly_global_active_power = df['Global_active_power'].resample('M').mean()

df_monthly_global_active_power_pd = df_monthly_global_active_power.compute()

plt.figure(figsize=(12, 6))

df_monthly_global_active_power_pd.plot(title='Promedio mensual de Global Active Power', color='tab:blue')
plt.ylabel('Global Active Power (kW)')
plt.xlabel('Fecha')
plt.grid(True)
plt.show()
```

```python
# Promedio mensual de Global Reactive Power (en kW)
df_monthly_global_reactive_power = df['Global_reactive_power'].resample('M').mean().compute()
plt.figure(figsize=(12, 6))
df_monthly_global_reactive_power.plot(title='Promedio mensual de Global Reactive Power', color='tab:green')
plt.ylabel('Global Reactive Power (kW)')
plt.xlabel('Fecha')
plt.grid(True)
plt.show()
```

```python
# Promedio mensual de Voltage (en V)
df_monthly_voltage = df['Voltage'].resample('M').mean().compute()
plt.figure(figsize=(12, 6))
df_monthly_voltage.plot(title='Promedio mensual de Voltage', color='tab:red')
plt.ylabel('Voltage (V)')
plt.xlabel('Fecha')
plt.grid(True)
plt.show()
```

```python
# Promedio mensual de Global Intensity (en A)
df_monthly_global_intensity = df['Global_intensity'].resample('M').mean().compute()
plt.figure(figsize=(12, 6))
df_monthly_global_intensity.plot(title='Promedio mensual de Global Intensity', color='tab:orange')
plt.ylabel('Global Intensity (A)')
plt.xlabel('Fecha')
plt.grid(True)
plt.show()
```

```python
df_cleaned_exploratory['Global_active_power'] = df_cleaned_exploratory['Global_active_power'] * (1000 / 60)
df_cleaned_exploratory['Global_reactive_power'] = df_cleaned_exploratory['Global_reactive_power'] * (1000 / 60)

# Cálculo de active_energy_consumed
df_cleaned_exploratory['active_energy_consumed'] = (df_cleaned_exploratory['Global_active_power'] -
                                       (df_cleaned_exploratory['Sub_metering_1'] +
                                        df_cleaned_exploratory['Sub_metering_2'] +
                                        df_cleaned_exploratory['Sub_metering_3']))

# Variables de interés y unidades
variables = {
    'Global_active_power': 'Wh',
    'Global_reactive_power': 'Wh',
    'Voltage': 'V',
    'Sub_metering_1': 'Wh',
    'Sub_metering_2': 'Wh',
    'Sub_metering_3': 'Wh',
    'active_energy_consumed': 'Wh'
}

# Función para calcular resumen estadístico personalizado
def resumen_estadistico(df, freq):
    stats_data = []
    # Resample según la frecuencia (T=minuto, D=día, W=semana, M=mes)
    df_resampled = df.resample(freq).mean()

    for var, unit in variables.items():
        max_value = df_resampled[var].max()
        min_value = df_resampled[var].min()
        mean_value = df_resampled[var].mean()
        std_value = df_resampled[var].std()
        median_value = df_resampled[var].median()

        stats_data.append({
            'Tipo de Variable': var,
            'Unidad': unit,
            'Frecuencia': freq,
            'Valor Máximo': round(max_value, 3),
            'Valor Mínimo': round(min_value, 3),
            'Valor Medio': round(mean_value, 3),
            'Desviación Estándar': round(std_value, 3),
            'Mediana': round(median_value, 3)
        })

    return pd.DataFrame(stats_data)

# Resúmenes por minuto, día, semana y mes
resumen_minuto = resumen_estadistico(df_cleaned_exploratory, 'T')  # Minuto
resumen_dia = resumen_estadistico(df_cleaned_exploratory, 'D')     # Día
resumen_semana = resumen_estadistico(df_cleaned_exploratory, 'W')  # Semana
resumen_mes = resumen_estadistico(df_cleaned_exploratory, 'M')     # Mes

# Concatenar todos los resúmenes
resumen_total = pd.concat([resumen_minuto, resumen_dia, resumen_semana, resumen_mes])

print(tabulate(resumen_total, headers='keys', tablefmt='grid'))

```

### LSTM

#### Carga y configuracion de cluster de Dask

Configuramos la conexión a un clúster Dask para procesar los datos de forma distribuida

```python
# Configurar Cliente Dask para Paralelización**
client = Client(n_workers=2, threads_per_worker=2, memory_limit='12GB')  # Ajusta según tus recursos

# Verificar la GPU
print("GPU disponible:", tf.config.list_physical_devices('GPU'))

# Cargar y Preparar Datos
file_path = 'household_power_consumption.txt'

# Leer datos con Dask y filtrar columnas
df_nulos = dd.read_csv(
    file_path, sep=';', na_values=['?'], usecols=['Date', 'Time', 'Global_active_power']
)

# Convertir columnas de fecha y tiempo a datetime
df_nulos['DateTime'] = dd.to_datetime(
    df_nulos['Date'] + ' ' + df_nulos['Time'], format='%d/%m/%Y %H:%M:%S'
)

# Eliminar valores nulos
df_cleaned = df_nulos.dropna(subset=['Global_active_power']).drop(['Date', 'Time'], axis=1).persist()

# Asegurar que el índice esté ordenado y sin nulos
df_cleaned = df_cleaned.set_index('DateTime', sorted=True)

# Reamostrar a datos diarios y calcular la media
df_daily = df_cleaned.resample('D').mean().dropna()
```

#### Normalizacion de los datos

```python
# Normalizar los datos con MinMaxScaler
scaler = MinMaxScaler()

# Normalizar los valores y convertirlos a un DataFrame de Pandas
scaled_values = scaler.fit_transform(df_daily[['Global_active_power']].compute())

# Convertir el array normalizado en un DataFrame de Dask
scaled_df = dd.from_array(scaled_values, columns=['Global_active_power'])

# Combinar el DataFrame normalizado con el resto de las columnas
df_daily = df_daily.drop('Global_active_power', axis=1)  # Eliminar columna original
df_daily = dd.concat([df_daily, scaled_df], axis=1)  # Combinar DataFrame original con el normalizado
```

#### Crear secuencia X, y

```python

# Crear secuencias para la serie temporal
def create_sequences(data, seq_length):
    X = np.array([data[i:i + seq_length] for i in range(len(data) - seq_length)])
    y = np.array([data[i + seq_length] for i in range(len(data) - seq_length)])
    return X, y

seq_length = 4
data = df_daily['Global_active_power'].values.compute()
X, y = create_sequences(data, seq_length)
```

#### Partición y preparción en Tensorflow

Se hace la partición en 80% para entrenamiento y 20% para test, ya que posteriormente se empleará el parámetro "Validation_split" con un valor de 0.2, haciendo el proceso de entrenamiento, validación y test.

```python
# Dividir en entrenamiento y prueba
train_size = int(0.8 * len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Preparar para TensorFlow
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

#### Función de modelado y evaluación

```python
def create_gru_model(units, learning_rate, input_shape):
    model = tf.keras.Sequential([
        tf.keras.layers.GRU(units, activation='relu', input_shape=input_shape),
        tf.keras.layers.Dropout(0.2),  # Regularización
        tf.keras.layers.Dense(1)
    ])
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    # def log_cosh_loss(y_true, y_pred):
      # return tf.reduce_mean(tf.math.log(tf.cosh(y_pred - y_true)))

    # model.compile(optimizer=optimizer, loss=log_cosh_loss)
    model.compile(optimizer=optimizer, loss=tf.keras.losses.Huber())
    return model

def evaluate_model(X_train, y_train, X_test, y_test, units, batch_size, learning_rate, epochs=10, scaler=None):
    input_shape = (X_train.shape[1], X_train.shape[2])
    model = create_gru_model(units, learning_rate, input_shape)
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)


    model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.2,
              callbacks=[early_stopping], verbose=0)

    y_pred = model.predict(X_test)

    if scaler is not None:
        y_pred_inv = scaler.inverse_transform(y_pred)
        y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))
        rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
    else:
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    return rmse
```

#### Optimizacion del modelo y Entranamiento

```python
param_dist = {
    'units': [30, 40, 50],
    'batch_size': [22, 32, 64],
    'learning_rate': [0.001, 0.005, 0.0005],
    'epochs': [40, 50, 60, 70]
}

# Seleccionar combinaciones aleatorias
num_samples = 30  # Número de combinaciones a probar
random_params = list(ParameterSampler(param_dist, n_iter=num_samples, random_state=42))

best_params = None
best_rmse = float("inf")

print("\nIniciando Random Search...")
for params in random_params:
    print(f"Probando parámetros: {params}")
    rmse = evaluate_model(
        X_train, y_train, X_test, y_test,
        units=params['units'],
        batch_size=params['batch_size'],
        learning_rate=params['learning_rate'],
        epochs=params['epochs'],
        scaler=scaler
    )
    print(f"RMSE: {rmse:.4f}")
    if rmse < best_rmse:
        best_rmse = rmse
        best_params = params

print("\nMejores Parámetros:")
print(best_params)
print(f"Mejor RMSE: {best_rmse:.4f}")

# **6. Entrenamiento Final con Mejores Parámetros**
print("\nEntrenamiento Final con Mejores Parámetros...")
final_model = create_gru_model(
    units=best_params['units'],
    learning_rate=best_params['learning_rate'],
    input_shape=(X_train.shape[1], X_train.shape[2])
)
final_model.fit(
    X_train, y_train,
    epochs=best_params['epochs'],
    batch_size=best_params['batch_size'],
    validation_split=0.2,
    verbose=1
)

# Predicción Final
y_pred = final_model.predict(X_test)
y_pred_inv = scaler.inverse_transform(y_pred)
y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))

# Evaluación Final
rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
mae = mean_absolute_error(y_test_inv, y_pred_inv)
# Calcular R²
r2 = r2_score(y_test_inv, y_pred_inv)
print("\nEvaluación Final del Modelo:")
print(f"RMSE: {rmse:.3f}")
print(f"MAE: {mae:.3f}")
print(f"R²: {r2:.3f}")

# Visualización
plt.figure(figsize=(10, 6))
plt.plot(y_test_inv, label='Real', color='blue')
plt.plot(y_pred_inv, label='Pronóstico', color='red')
plt.title('GRU: Predicción Optimizada de Consumo')
plt.xlabel('Tiempo')
plt.ylabel('Global Active Power')
plt.legend()
plt.show()
```

**Inferencias:**

* RMSE y MAE: Indican que el modelo tiene una capacidad razonable para predecir valores de consumo, aunque los errores absolutos aún son relativamente altos en términos de la escala de los datos.

* R²: Un valor de 0.398 sugiere que el modelo explica cerca del 39.8% de la variabilidad de los datos reales. Esto indica que hay espacio para mejoras, ya que el modelo no captura toda la complejidad de la relación entre las variables.

**Comparacion visual de Predicciones vs. Valores Reales:**

Tendencias Generales: El modelo logra seguir las tendencias generales del consumo energético, especialmente en las regiones donde los valores fluctúan de manera moderada.

Limitaciones:

* Desempeño en picos extremos: Los valores reales presentan picos (altos y bajos) más pronunciados que el modelo no logra captar.

* Los mejores parámetros encontrados (units=40, learning_rate=0.001, epochs=50, batch_size=22) indican que:
  * Un número moderado de unidades ayuda a capturar patrones sin sobreajuste.
  * El aprendizaje más lento (tasa de 0.001) permite convergencia estable.
  * El tamaño de batch pequeño mejora la representación de los datos en cada iteración.

### **Clustering**

De manera análoga al modelo anterior se desea implementar una herramienta que agrupe horas de consumo y electrodomesticos, de esta manera, se podrán elaborar estrategias para un uso responsable de las herramientas en el hogar.

Este enfoque consiste en seleccionar subconjuntos de variables que permiten analizar patrones de consumo específicos en diferentes contextos. En este caso, podemos centrarnos en las siguientes combinaciones:

Sub_metering_1, Sub_metering_2 y Sub_metering_3: Representan el consumo específico en cocina, lavandería y calentador/aire acondicionado.
Global_reactive_power y Voltage: Analiza la relación entre potencia reactiva y voltaje para detectar ineficiencias en el sistema eléctrico.
Global_active_power: Relaciona la corriente con la potencia activa, útil para analizar la carga total en el hogar.

```python
# Configurar el cliente Dask
client = Client(n_workers=2, threads_per_worker=2, memory_limit='12GB')

# Cargar el dataset con Dask
file_path = 'household_power_consumption.txt'
df_dask = dd.read_csv(file_path, sep=';', na_values=['?'], low_memory=False)

# Eliminar valores nulos
df_dask = df_dask.dropna()

# Asegurar que las columnas relevantes sean numéricas
columns_to_convert = ['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3',
                      'Global_reactive_power', 'Voltage', 'Global_active_power']
df_dask[columns_to_convert] = df_dask[columns_to_convert].astype(float)

# Crear columna de fecha y hora para reamostrar
df_dask['DateTime'] = dd.to_datetime(df_dask['Date'] + ' ' + df_dask['Time'], format='%d/%m/%Y %H:%M:%S')
df_dask = df_dask.set_index('DateTime')

# Ordenar índice y establecer divisiones conocidas
df_dask = df_dask.repartition(npartitions=10).persist()

# Variables seleccionadas para el clustering
selected_features = {
    'Sub_metering': ['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3'],
    'Reactive_Voltage': ['Global_reactive_power', 'Voltage'],
    'Power_Intensity': ['Global_active_power']
}

# Diccionario para guardar resultados
cluster_results = {}

# Búsqueda aleatoria de hiperparámetros
param_dist = {
    'n_clusters': [2, 3, 4, 5],  # Número de clusters
    'init': ['random'],  # Cambiar a 'random' para evitar el error con 'k-means++'
    'max_iter': [100, 300]  # Iteraciones máximas
}

# Número de combinaciones aleatorias a probar
num_samples = 5

for name, features in selected_features.items():
    print(f"\nClustering basado en: {name}")

    # Seleccionar las características y reamostrar a nivel diario
    df_selected = df_dask[features].compute()  # Computar el DataFrame
    df_selected = df_selected.resample('D').mean().dropna()  # Reamostrar

    # Normalización de los datos con Dask-ML
    scaler = MinMaxScaler()
    df_scaled = scaler.fit_transform(df_selected)

    # Configuración de parámetros para búsqueda aleatoria
    random_params = list(ParameterSampler(param_dist, n_iter=num_samples, random_state=42))

    best_params = None
    best_inertia = float("inf")

    print("\nIniciando búsqueda aleatoria de hiperparámetros...")
    for params in random_params:
        print(f"Probando parámetros: {params}")

        # Entrenar modelo KMeans con parámetros actuales
        kmeans = KMeans(
            n_clusters=params['n_clusters'],
            init=params['init'],
            max_iter=params['max_iter'],
            random_state=42
        )
        kmeans.fit(df_scaled)

        # Evaluar la inercia
        inertia = kmeans.inertia_
        print(f"Inercia: {inertia:.4f}")

        # Guardar los mejores parámetros
        if inertia < best_inertia:
            best_inertia = inertia
            best_params = params

    print(f"\nMejores parámetros para {name}: {best_params}")
    print(f"Mejor inercia: {best_inertia:.4f}")

    # Entrenar modelo final con los mejores parámetros
    kmeans = KMeans(
        n_clusters=best_params['n_clusters'],
        init=best_params['init'],
        max_iter=best_params['max_iter'],
        random_state=42
    )
    kmeans.fit(df_scaled)

    # Obtener predicciones de clusters
    clusters = kmeans.predict(df_scaled)

    # Convertir los clusters en una serie y asignarlos al DataFrame
    df_selected['Cluster'] = clusters

    # Guardar resultados en el diccionario
    cluster_results[name] = df_selected

    # Visualización del clustering
    plt.figure(figsize=(10, 6))
    plt.scatter(df_selected.index, df_selected[features[0]], c=df_selected['Cluster'], cmap='viridis', alpha=0.7)
    plt.title(f'Clustering basado en {name}')
    plt.xlabel('Fecha')
    plt.ylabel(features[0])
    plt.colorbar(label='Cluster')
    plt.show()
```

Análisis:

* Clustering basado en Power_Intensity:
  * Mejor inercia: 44.4083 con n_clusters=5.
  * Los datos muestran patrones claros en el consumo de energía activa, y los clusters representan diferentes niveles de intensidad de consumo.
  * Los clusters están bien diferenciados, con patrones que coinciden con picos y valles en el consumo de energía activa.
  * Las variaciones estacionales parecen ser capturadas en algunos clusters.

* Clustering basado en Reactive_Voltage:
  * Mejor inercia: 114.9778 con n_clusters=5.
  * Los clusters aquí parecen capturar patrones en la relación entre el poder reactivo y el voltaje. Esto puede reflejar diferentes patrones de comportamiento en la red eléctrica.
  * Los clusters reflejan variaciones en el uso de submediciones. Los datos son más dispersos, lo que sugiere patrones más complejos.

* Clustering basado en Sub_metering:

  * Mejor inercia: 201.5289 con n_clusters=5.
  * Los clusters aquí representan variaciones en el consumo de submediciones. Los datos son más dispersos, pero aún se pueden identificar grupos significativos.
  * Los clusters capturan cambios en la combinación de poder reactivo y voltaje. Esto podría estar relacionado con cambios en la carga de la red o picos específicos en el uso.
